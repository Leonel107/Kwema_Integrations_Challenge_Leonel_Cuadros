```python
# Libraries
import requests as rq
import json as js
from bs4 import BeautifulSoup
import pandas as pd
import numpy as n

# Enter Github username
username = input("Enter Github username: ")
print(username)
```

### 1. Extract programming languages from repositories

```python
# extract the programming languages used in the user's repository
page = rq.get("https://github.com/{}?tab=repositories".format(username))
page.status_code

text_soup = BeautifulSoup(page.text,'lxml')

text_filter_programmingLanguage = text_soup.find_all('span',class_='ml-0 mr-3')
print(text_filter_programmingLanguage)
print(len(text_filter_programmingLanguage))
```

```python
# extract the programming languages used in the user's repository
programming_languages = []
for languages in text_filter_programmingLanguage:
    name_language = languages.find(itemprop='programmingLanguage')
    name_language = name_language.get_text()
    programming_languages.append(name_language)
    
print(programming_languages)
print(len(programming_languages))
```

### 2. Pull the repositories with the highest acceptance of Pull Request requests

```python
# Extract information from the repository
page = rq.get("https://github.com/{}?tab=repositories".format(username))
page.status_code

text_soup = BeautifulSoup(page.text,'lxml')

text_filter_repositories = text_soup.find_all('div',class_='col-10 col-lg-9 d-inline-block')
print(text_filter_repositories)
print(len(text_filter_repositories))
```

### 2.1. EXTRACT REPOSITORY NAMES

```python
# Extract Rrpository Names
repositories_name = []
for repositories in text_filter_repositories:
    name_repository = repositories.find('a',itemprop='name codeRepository')
    name_repository = name_repository.get_text().strip()
    repositories_name.append(name_repository)
    
print(repositories_name)
print(len(repositories_name))
```

### 2.2. EXTRACT TOTAL PAGES FROM THE REPOSITORY

```python
# Get the number of pages from the pull request
page = rq.get("https://github.com/{}/linux/pulls?q=is%3Apr+is%3Aclosed".format(username))
page.status_code

text_soup = BeautifulSoup(page.text,'lxml')
maximo_pagina = int(text_soup.em['data-total-pages'])

text_filter_number_pages = text_soup.find_all(class_="current")
print(text_filter_number_pages)
print(len(text_filter_number_pages))
print(maximo_pagina)
```

### 2.3. EXTRACT THE REPOSITORIES WITH THE HIGHEST ACCEPTED PULL REQUEST

```python
# Pull the repositories with the highest pull request acceptance
list_pull_request_count = [] # Repository pull request counter

for j in repositories_name:
    pull_request_accepted = [] # Repository pull counter
    for i in range(1, maximo_pagina + 1):
        page = rq.get("https://github.com/{}/{}/pulls?page={}&q=is%3Apr+is%3Aclosed".format(username, j, i))
        page.status_code
    
        text_soup = BeautifulSoup(page.text,'lxml')
    
        text_filter_pull_request = text_soup.find_all('div',class_="Box-row Box-row--focus-gray p-0 mt-0 js-navigation-item js-issue-row")
    
    #print(text_filter_pull_request)
    #print(len(text_filter_pull_request))
        
    # Pull the repositories with the highest pull request acceptance
    
        for pull_request in text_filter_pull_request:
            pull_request_ok = pull_request.find('svg', class_="octicon octicon-git-merge color-fg-done")
            if pull_request_ok != None:
                pull_request_ok = pull_request_ok.get_text().strip()
                pull_request_accepted.append(pull_request_ok)

    list_pull_request_count.append(len(pull_request_accepted))

print(list_pull_request_count)
print(len(list_pull_request_count))
```

### DATAFRAME CREATION

```python
# Dataframe
df_username = pd.DataFrame()
df_username['Repository Name'] = repositories_name
df_username['Programming_Languages'] = programming_languages
df_username['Pull Request Accepted'] = list_pull_request_count
df_username
```
